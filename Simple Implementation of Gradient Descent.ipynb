# Gradient Descent

```python
#importing neccesary libraries
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.array ([0.5,2.3,2.9])
y= np.array([1.4,1.9,3.2])
```

```python
plt.figure(figsize =(12,7))
plt.title('X and Y graph' ,fontsize=24)
plt.scatter(x,y,s=256 , alpha =1 ,label ='Y')
```

```python
b0,b1 =0.0 , 0.65
lr = 0.001
epochs = 1000
error = []
```

```python
for epoch in range(epochs):
    epoch_cost , cost_b0,cost_b1 = 0,0,0
    for i in range(len(x)):
        #make the prediction
        y_pred = (b0 + b1*x[i])
        # append the squared error
        epoch_cost+= (y[i] - y_pred)**2
        
        for j in range(len(x)):
            #partial diff of b0,b1 for current row then add them to find the total
            partial_b0 = -2 * (y[j] - (b0 + b1*x[j]))
            partial_b1 = (-2 * x[j]) * (y[j] - (b0 + b1*x[j]))
            
            cost_b0+= partial_b0
            cost_b1+= partial_b1
            
        b0 = b0 - lr*cost_b0
        b1 = b1 - lr*cost_b1
        
    error.append(epoch_cost)
           
```

```python
plt.figure(figsize=(12,7))
plt.title('Gradient Descent- error over 1000 epochs' , fontsize= 24)
plt.plot(np.arange(1,len(error)+1) ,error )
```
